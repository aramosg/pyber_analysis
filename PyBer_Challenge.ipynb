### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
# In Step 1, use the groupby() function to create a Series of data 
# that has the type of city as the index, then apply the count() method to the "ride_id" column.
#---
#  1. Get the total rides for each city type

# Este query genera un DF con el # de rides por tipo de ciudad
total_rides_per_city_type = pyber_data_df.groupby("type").count()["ride_id"]

# Query individual por type of city
#urban_total_rides = pyber_data_df[pyber_data_df["type"] == "Urban"].count()["ride_id"]

total_rides_per_city_type # -- GOOD!

#pyber_data_df[pyber_data_df["type"] == "Suburban"]
```

```python
#In Step 2, use the groupby() function to create a Series of data that has the 
#type of city as the index, then apply the sum() method to the "driver_count" column.
# ---
# 2. Get the total drivers for each city type
total_drivers_per_city_type = city_data_df.groupby("type").sum()["driver_count"]

total_drivers_per_city_type # GOOD!
```

```python
#In Step 3, use the groupby() function to create a Series of data that has the 
#type of city as the index, then apply the sum() method to the "fare" column.
#---
#  3. Get the total amount of fares for each city type
total_fares_per_city_type = pyber_data_df.groupby("type").sum()["fare"]
total_fares_per_city_type # -- GOOD!
```

```python
#In Step 4, calculate the average fare per ride by city type by 
#dividing the sum of all the fares by the total rides.
#--------
#  4. Get the average fare per ride for each city type. 
mean_fare_per_city_type = pyber_data_df.groupby(["type"]).mean()["fare"]
mean_fare_per_city_type # GOOD!
```

```python
#In Step 5, calculate the average fare per driver by city type by dividing 
#the sum of all the fares by the total drivers.
#---
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares_per_city_type / total_drivers_per_city_type
average_fare_per_driver #GOOD!
```

```python
#In Step 6, create a PyBer summary DataFrame with all the data gathered
#from Steps 1-5, using the column names shown below:.
#---
#  6. Create a PyBer summary DataFrame. 
dictionary = {
    "Total Rides":total_rides_per_city_type,
    "Total Drivers": total_drivers_per_city_type,
    "Total Fares": total_fares_per_city_type,
    "Average Fare per Ride": mean_fare_per_city_type,
    "Average Fare per Driver": average_fare_per_driver
}
pyber_summary_df = pd.DataFrame(dictionary)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
#district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
#district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
#district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
group_by_type_and_date_df = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
group_by_type_and_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
group_by_type_and_date_df = group_by_type_and_date_df.reset_index()
group_by_type_and_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_df = group_by_type_and_date_df.pivot(index="date", columns="type", values="fare")
pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
dates_df = pivot_df.loc["2019-01-01":"2019-04-29"]
dates_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
dates_df.index = pd.to_datetime(dates_df.index)
#dates_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
dates_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resample_df = dates_df.resample("W").sum()
resample_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.dates as mdates # -- THIS IS NEW
import datetime # -- This is new

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=((16,6)))
ax.plot(resample_df, label=list(resample_df.columns)) # Adding data to the plot
ax.set_title("Total Fare by City Type", fontsize=18) # Setting title
ax.set_ylabel("Fare ($USD)", fontsize=12) # Setting Y label
ax.set_xlabel("2019", fontsize=12, loc="left") # Setting X label

list_for_x_ticks = [] # dinamically adding the x ticks lables

for ind in resample_df.index:
    the_year = ind.year
    the_month = ind.month
    fecha = datetime.datetime(the_year, the_month, 1)
    if fecha not in list_for_x_ticks:
        list_for_x_ticks.append(fecha)

ax.set_xticks(list_for_x_ticks) # adding the x ticks
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y')) # Formatting the x ticks - month and year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b')) # Formatting the x ticks - month only

# Creating the legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="center", title="Type")
#lgnd.legendHandles[0]._sizes = [75]
#lgnd.legendHandles[1]._sizes = [75]
#lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Saving the plot to a file
plt.savefig("analysis/PyBer_Challenge_final_plot.png")

plt.show()

```
